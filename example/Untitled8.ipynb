```python
# encoding: utf-8

import pandas as pd
from pandas import *
import datetime
import json
from pymongo import MongoClient
from collections import defaultdict

pd.set_option('display.width', None)  # 设置字符显示宽度
pd.set_option('display.max_rows', None)  # 设置显示最大行
pd.set_option('display.max_columns', None)  # 设置显示最大行

client = MongoClient('localhost', 27017)
db = client.futures
indexMarket = db.indexMarket
unit=db.unit
peak2=db.peak2

start='20190601'
# var='JD'

indexMarket = DataFrame(list(indexMarket.find({'date': {'$gte': start}})))
unit = DataFrame(list(unit.find()))
dd=unit['variety']


df2={}
for i in dd:
    try:
        df=indexMarket[indexMarket['variety']==i]

# df
        df= df[['date', 'variety','set_close', 'set_high', 'set_low']].drop_duplicates()

        df['date'] = pd.to_datetime(df['date'])
        df.set_index('date',inplace=True)
        maxs = df.rolling(window=10, on='variety').max().dropna()
        mins = df.rolling(window=10, on='variety').min().dropna()
        hb = pd.merge(maxs, mins, on=['date', 'variety'], how='outer')
        hb['gains']=round(hb.apply(lambda x: (x['set_high_x'] / x['set_low_y']-1)*100, axis=1), 2)
# hb
# #         # 跌幅
        hb['lesses'] = round(hb.apply(lambda x:(1-x['set_low_y'] / x['set_high_x'])*100,axis=1), 2)
        hb.tail(2)
        hb['peak']=hb.apply(lambda x:x['gains'] if x['set_low_y'] < x['set_close_x']else x['lesses'], axis=1)
#         hb['variety']=i
#         print(hb.tail(3))
        hb = hb.reset_index()
#         print(hb)
#         hb=pd.DataFrame(hb)
        peak2.insert_one(json.loads(hb.T.to_json()).values())
        print(json.loads(hb.T.to_json()).values())

    except:
#         pass
        print('??')
#     continue
```
