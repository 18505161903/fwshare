```python
import pandas as pd
import numpy as np

aa = np.array([1, 0, 1, 0])
bb = pd.DataFrame(aa.T, columns=['one'])
# 生成一个ndarray，装要插入的值
two = np.zeros(bb.shape[0])
# 按条件修改two
print(two)
for i in range(bb.shape[0]):
    if bb['one'][i] == 0:
        two[i] = 1
# 完成后将two插入dataframe中
bb.insert(1,'two', two)
print(bb)
```
